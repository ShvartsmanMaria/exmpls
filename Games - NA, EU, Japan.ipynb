Пропуски в **оценках и рейтинге**:

```python
critic_missing = data['critic_score'].isna().sum()
critic_missing_rate = critic_missing / len(data) * 100
print("Процент отсутствующих данных в столбце critic_score:", round(critic_missing_rate, 2))

user_missing = data['user_score'].isna().sum()
user_missing_rate = user_missing / len(data) * 100
print("Процент отсутствующих данных в столбце user_score:", round(user_missing_rate, 2))

rating_missing = data['rating'].isna().sum()
rating_missing_rate = rating_missing / len(data) * 100
print("Процент отсутствующих данных в столбце rating:", round(rating_missing_rate, 2))
```

Попробуем увидеть, есть ли **устойчивые паттерны в пропусках** (заодно это поможет сделать обоснованные предположения о причинах отсутствия данных): 

```python
# посмотрим для примера на critic_score (больше всего пропусков):
display(data[data['critic_score'].isnull()].head(20))
```

Видим, что без рейтингов и оценок у нас остались игры разных жанров, на разных платформах, в том числе и с относительно хорошими продажами. По году релиза тоже не видно закономерностей: есть и старые игры, и новые (с учетом того, что у нас сейчас 2016 год :)<br>
Однако четко прослеживается закономерность: там, где нет оценок критиков, там чаще всего и пользовательских оценок нет, и рейтинга. 
Посмотрим внимательнее: 

```python
compare_score_nans1 = len(data[data['critic_score'].isnull() & data['user_score'].isnull()])
compare_score_nans_rate1 = compare_score_nans1 /len(data[data['critic_score'].isnull()]) * 100
print("Совпадение пропусков в critic_score и user_score по отношению ко всем пропускам в critic_score:", round(compare_score_nans_rate1, 2))

compare_score_nans2 = len(data[data['critic_score'].isnull() & data['rating'].isnull()])
compare_score_nans_rate2 = compare_score_nans2 /len(data[data['critic_score'].isnull()]) * 100
print("Совпадение пропусков в critic_score и rating по отношению ко всем пропускам в critic_score:", round(compare_score_nans_rate2, 2))


compare_score_nans3 = len(data[data['user_score'].isnull() & data['rating'].isnull()])
compare_score_nans_rate3 = compare_score_nans3 /len(data[data['user_score'].isnull()]) * 100
print("Совпадение пропусков в user_score и rating по отношению ко всем пропускам в user_score:", round(compare_score_nans_rate3, 2))
```

Видно, что пропуски в столбцах critic_score, user_score и rating в значительной степени совпадают. <br>

Вот как выглядят совпадения в пропусках:

Стоит также проверить, есть ли зависимость между пропусками в **оценках и рейтинге и платформами**:

```python
# смотрим на примере critic_score:
critic_platform_rel = data.groupby("platform")['critic_score']
display(critic_platform_rel.count() / critic_platform_rel.size() * 100)
```

По многим платформам вообще нет данных об оценках и рейтинге. Однако и по тем платформам, по которым данные существуют, всегда есть значительное количество пропусков. 

Неопровергнутыми остались только два **предположения о причинах пропусков в оценках и рейтинге**: <br>
1. в таблицу сведены данные из разных источников, не все их которых в принципе содержат информацию об оценках и рейтингах; <br>
2. в разных странах действуют разные системы оценок и рейтингов, их не попытались категоризировать, унифицировать и так далее, а просто взяли данные только по одной из таких систем, остальные - выкинули.

Логическим образом восстановить пропущенные данные в оценках критиков, пользователей и в рейтинге невозможно, поэтому при анализе будем использовать только исходные данные. <br>
NaN-значения в числовых столбцах оставим на месте, в рейтинге - меняем на маркер, чтобы сохранить единый формат данных в столбце.

```python
data['rating'].fillna("NaN", inplace = True)
data['rating'].unique()
```

tbd - 'to be defined' в данном случае не несёт никого смысла: большая часть из них относится к давно вышедшим играм, - если там все еще нет оценок, они уже не появятся. По сути, для нас это отсутствующие значения. 
Заменим tbd на NaN приводим всё к float64.

```python
data['user_score'].replace('tbd', np.nan, inplace = True)
data['user_score'] = pd.to_numeric(data['user_score']).astype(float)
assert data.dtypes['user_score'] == 'float64'
```

### Влияние на продажи внутри одной популярной платформы отзывов пользователей и критиков. Диаграмма рассеяния и корреляция между отзывами и продажами.  

Рассмотрим для примера PS4 за актуальный период: по ней больше всего данных. 

```python
correlation_set = data_actual[['platform', 'total_sales', 'critic_score', 'user_score']]
correlation_set.query('platform == "PS4"').plot(x = "critic_score", y = 'total_sales', kind = "scatter", alpha = 0.5)
plt.show()
```

```python
correlation_set.query('platform == "PS4" and user_score > 0').plot(x = "user_score", y = 'total_sales', kind = "scatter", alpha = 0.5)
plt.show()
```

Похоже, хорошие отзывы критиков и пользователей не гарантируют хороших продаж (зато очень низкий рейтинг  - гарантирует плохие). <br>
В целом, корреляция между оценками критиков и продажами прослеживается гораздо чётче. Надо проверить цифрами:

```python
correlation_ps4 = correlation_set.query('platform == "PS4" and user_score > 0')
display(correlation_ps4['total_sales'].corr(correlation_ps4['critic_score']))
display(correlation_ps4['total_sales'].corr(correlation_ps4['user_score']))
```

Положительная корреляция между оценками критиков и продажами действительно есть, хотя и не высокая. <br>
Между оценками пользователей и продажами - нет. <br>
Диаграмма рассеяния иллюстрирует корреляцию:

```python
pd.plotting.scatter_matrix(correlation_ps4, figsize = (12, 12), alpha = 0.3, color = 'red')
plt.show()
```

```python
# и то же в числах:
correlation_ps4.corr()
```

Итак, на отзывы критиков очень приблизительно, но полагаться можно. А на пользователей - нет. <br>
Стоит учесть, что **correlation does not imply causation**. У нас нет оснований для выводов, что отзывы критиков именно влияют на продажи (как этот вопрос был сформулирован в задании к проекту) - для этого нам нужны другие данные. Но предположение, что **оценки критиков могут служить маркером, показывающим коммерческий потенциал игры**, кажется обоснованным.

## Проверка гипотез.

### Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и альтернативную гипотезы:<br>
**H<sub>0</sub>**: Средние пользовательские рейтинги платформ Xbox One и PC равны.<br>
**H<sub>1</sub>**: Средние пользовательские рейтинги платформ Xbox One и PC не равны.

Создадим переменные, чтобы удобней было обращаться к нужным данным, заодно отфильтруем маркеры в оценках пользователей:

```python
scores_xbox = data_actual.query('user_score > 0')
scores_xbox = scores_xbox[scores_xbox['platform'] == 'XOne']['user_score']
scores_pc = data_actual.query('user_score >0')
scores_pc = scores_pc[scores_pc['platform'] == 'PC']['user_score']
```

Для проверки гипотезы используем тест Стьюдента (t-test):

```python
# задаем критический уровень статистической значимости:
alpha = 0.05

# проверяем гипотезу о равенстве двух генеральных совокупностей по взятым из них выборкам:
result = st.ttest_ind(scores_xbox, scores_pc)

print('р-значение:', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу.')
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

С вероятностью 10% различия в пользовательских рейтингах между Xbox One и PC случайно и на самом деле средние рейтинги равны. **Гипотеза о равенстве пользовательских рейтингов XBox One и PC не отвергнута".**

А вот и рейтинги:

```python
data_actual.query('user_score > 0').pivot_table(index = "platform", values = "user_score", aggfunc = "mean")\
.sort_values(by = "user_score", ascending = False)
```

### Проверка гипотезы: средние пользовательские рейтинги жанров Action и Sports разные

Сформулируем нулевую и альтернативную гипотезы:<br>
**H<sub>0</sub>**: Средние пользовательские рейтинги жанров Action и Sports равны.<br>
**H<sub>1</sub>**: Средние пользовательские рейтинги платформ Xbox One и PC не равны.

```python
scores_action = data_actual.query('user_score > 0')
scores_action = scores_action[scores_action['genre'] == 'Action']['user_score']
scores_sports = data_actual.query('user_score >0')
scores_sports = scores_sports[scores_sports['genre'] == 'Sports']['user_score']
```

```python
# задаем критический уровень статистической значимости:
alpha = 0.05

# проверяем гипотезу о равенстве двух генеральных совокупностей по взятым из них выборкам:
result = st.ttest_ind(scores_action, scores_sports)

print('р-значение:', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу.')
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Таким образом, анализ выборок с помощью теста Стьюдента показывает, что вероятность того, что нулевая гипотеза (гипотеза о равенстве средних пользовательских рейтингов жанров Action и Sports) верна - ничтожно мала, что **подтверждает альтернативную гипотезу: средние пользовательские рейтинги жанров Action и Sports не равны**.

Нулевая гипотеза в обоих случаях сформулирована как предполагаемое равенство между выборками, потому что основное проверяемое с помощью теста Стьюдента предположение обычно формулируется как отсутствие различий, отсутствие влияния того или иного фактора, и т.п. 
Альтернативная гипотеза противоположна нулевой. <br>
